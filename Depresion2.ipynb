```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import pickle


Dd_data = pd.read_csv("Depresion.csv")


le_criterio1  = LabelEncoder()
Dd_data["Estado de ánimo deprimido la mayor parte del día, casi todos los días, según se desprende de la información subjetiva "] = le_criterio1.fit_transform(Dd_data["Estado de ánimo deprimido la mayor parte del día, casi todos los días, según se desprende de la información subjetiva "])
Dd_data["Estado de ánimo deprimido la mayor parte del día, casi todos los días, según se desprende de la información subjetiva "].unique()

le_criterio2 = LabelEncoder()
Dd_data["Disminución importante del interés o el placer por todas o casi todas las actividades la mayor parte del día, casi todos los días"] = le_criterio2.fit_transform(Dd_data["Disminución importante del interés o el placer por todas o casi todas las actividades la mayor parte del día, casi todos los días"])
Dd_data["Disminución importante del interés o el placer por todas o casi todas las actividades la mayor parte del día, casi todos los días"].unique()

le_criterio3 = LabelEncoder()
Dd_data["Insomnio o hipersomnia casi todos los días."] = le_criterio3.fit_transform(Dd_data["Insomnio o hipersomnia casi todos los días."])
Dd_data["Insomnio o hipersomnia casi todos los días."].unique()

le_criterio4 = LabelEncoder()
Dd_data["Agitación o retraso psicomotor casi todos los días"] = le_criterio4.fit_transform(Dd_data["Agitación o retraso psicomotor casi todos los días"])
Dd_data["Agitación o retraso psicomotor casi todos los días"].unique()
        
le_criterio5 = LabelEncoder()
Dd_data["Fatiga o pérdida de energía casi todos los días"] = le_criterio5.fit_transform(Dd_data["Fatiga o pérdida de energía casi todos los días"])
Dd_data["Fatiga o pérdida de energía casi todos los días"].unique()

le_criterio6 = LabelEncoder()
Dd_data["Sentimiento de inutilidad o culpabilidad excesiva o inapropiada casi todos los días "] = le_criterio6.fit_transform(Dd_data["Sentimiento de inutilidad o culpabilidad excesiva o inapropiada casi todos los días "])
Dd_data["Sentimiento de inutilidad o culpabilidad excesiva o inapropiada casi todos los días "].unique()  

le_criterio7 = LabelEncoder()
Dd_data["Disminución de la capacidad para pensar o concentrarse, o para tomar decisiones, casi todos los días "] = le_criterio7.fit_transform(Dd_data["Disminución de la capacidad para pensar o concentrarse, o para tomar decisiones, casi todos los días "])     
Dd_data["Disminución de la capacidad para pensar o concentrarse, o para tomar decisiones, casi todos los días "].unique()        

le_criterio8 = LabelEncoder()
Dd_data["Pensamientos de muerte recurrentes, ideas suicidas recurrentes sin un plan determinado, intento de suicidio o un plan específico para llevarlo a cabo."] = le_criterio8.fit_transform(Dd_data["Pensamientos de muerte recurrentes, ideas suicidas recurrentes sin un plan determinado, intento de suicidio o un plan específico para llevarlo a cabo."])        
Dd_data["Pensamientos de muerte recurrentes, ideas suicidas recurrentes sin un plan determinado, intento de suicidio o un plan específico para llevarlo a cabo."].unique()      

le_criterio9 = LabelEncoder()
Dd_data["Pérdida importante de peso sin hacer dieta o aumento de peso o disminución o aumento del apetito casi todos los días."] = le_criterio9.fit_transform(Dd_data["Pérdida importante de peso sin hacer dieta o aumento de peso o disminución o aumento del apetito casi todos los días."])        
Dd_data["Pérdida importante de peso sin hacer dieta o aumento de peso o disminución o aumento del apetito casi todos los días."].unique()        

le_criterio10 = LabelEncoder()
Dd_data["Nunca ha habido un episodio maníaco o hipomaníaco"] = le_criterio10.fit_transform(Dd_data["Nunca ha habido un episodio maníaco o hipomaníaco"])
Dd_data["Nunca ha habido un episodio maníaco o hipomaníaco"].unique()





x = Dd_data.drop(columns = ["Porcentaje"])
y = Dd_data["Porcentaje"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.5)
model = DecisionTreeClassifier()
model.fit(x, y)
predictions= model.predict(x_test)
score = accuracy_score(y_test, predictions)
data = {"model": model, "le_criterio1": le_criterio1, "le_criterio2": le_criterio2, "le_criterio3": le_criterio3, "le_criterio4": le_criterio4, "le_criterio5": le_criterio5, "le_criterio6": le_criterio6, "le_criterio7": le_criterio7, "le_criterio8": le_criterio8, "le_criterio9": le_criterio9, "le_criterio10": le_criterio10, }
with open('saved_stepsA.pkl', 'wb') as file:
    pickle.dump(data, file)
```
