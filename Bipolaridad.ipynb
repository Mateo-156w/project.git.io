```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import pickle

Bd_data = pd.read_csv("Bipolaridad.csv")






le_criterio1  = LabelEncoder()
Bd_data["Aumento de la autoestima o sentimiento de grandeza"] = le_criterio1.fit_transform(Bd_data["Aumento de la autoestima o sentimiento de grandeza"])
Bd_data["Aumento de la autoestima o sentimiento de grandeza"].unique()

le_criterio2 = LabelEncoder()
Bd_data["Disminución de la necesidad de dormir"] = le_criterio2.fit_transform(Bd_data["Disminución de la necesidad de dormir"])
Bd_data["Disminución de la necesidad de dormir"].unique()

le_criterio3 = LabelEncoder()
Bd_data["Más hablador de lo habitual o presión para mantener la conversación"] = le_criterio3.fit_transform(Bd_data["Más hablador de lo habitual o presión para mantener la conversación"])
Bd_data["Más hablador de lo habitual o presión para mantener la conversación"].unique()

le_criterio4 = LabelEncoder()
Bd_data["Fuga de ideas o experiencia subjetiva de que los pensamientos van a gran velocidad."] = le_criterio4.fit_transform(Bd_data["Fuga de ideas o experiencia subjetiva de que los pensamientos van a gran velocidad."])
Bd_data["Fuga de ideas o experiencia subjetiva de que los pensamientos van a gran velocidad."].unique()

le_criterio5 = LabelEncoder()
Bd_data["Facilidad de distracción"] = le_criterio5.fit_transform(Bd_data["Facilidad de distracción"])
Bd_data["Facilidad de distracción"].unique()

le_criterio6 = LabelEncoder()
Bd_data["Aumento de la actividad dirigida a un objetivo o agitación psicomotora "] = le_criterio6.fit_transform(Bd_data["Aumento de la actividad dirigida a un objetivo o agitación psicomotora "])
Bd_data["Aumento de la actividad dirigida a un objetivo o agitación psicomotora "].unique()

le_criterio7 = LabelEncoder()
Bd_data["Participación excesiva en actividades que tienen muchas posibilidades de consecuencias dolorosas"] = le_criterio7.fit_transform(Bd_data["Participación excesiva en actividades que tienen muchas posibilidades de consecuencias dolorosas"])     
Bd_data["Participación excesiva en actividades que tienen muchas posibilidades de consecuencias dolorosas"].unique() 

le_criterio8 = LabelEncoder()
Bd_data["Episodio de depresion mayor"] = le_criterio8.fit_transform(Bd_data["Episodio de depresion mayor"])
Bd_data["Episodio de depresion mayor"].unique() 




x = Bd_data.drop(columns = ["Porcentaje"])
y = Bd_data["Porcentaje"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.5)
model = DecisionTreeClassifier()
model.fit(x, y)
predictions= model.predict(x_test)
score = accuracy_score(y_test, predictions)

data = {"model": model, "le_criterio1": le_criterio1, "le_criterio2": le_criterio2, "le_criterio3": le_criterio3, "le_criterio4": le_criterio4, "le_criterio5": le_criterio5, "le_criterio6": le_criterio6, "le_criterio7": le_criterio7, "le_criterio8": le_criterio8,}
with open('saved_stepsG.pkl', 'wb') as file:
    pickle.dump(data, file)
 





```
